```python
def plot(df_list):
    
    fig, ax = plt.subplots()
    for i, e in enumerate(df_list):
        temp = e
        plt.plot(range(len(temp)), temp)
    # ax.legend()
    plt.legend()
    plt.show()
```

```python
def get_avg(runs, i):
    
    df_list = []
    for run in range(1, runs):
        temp = pd.read_csv('results/SE/results_' + str(run) + str(i) + '.csv')
        # temp = pd.read_csv('results/SE/final_data_' + str(run) + str(i) +'.csv')
        # temp = pd.read_csv('initial_data/initial_data_' + str(run) +'.csv')
        df_list.append(temp)
        # print(temp.shape)
    #     print(df_temp.columns)
    return sum(df_list) / len(df_list)
```

```python
def get_rolling_averaged_df(df, attr, n):
    
    new_attrs = []
    for a in attr:
        new_attr = 'rolAvg_' + a
        df[new_attr] = df[a].rolling(n).mean()
        df[new_attr].iloc[0] = df[a].iloc[0]
        new_attrs.append(new_attr)
    
    index = [x for x in range(df.shape[0]) if x%n == 0]
    temp_df = df.iloc[index]
    
    temp_df = temp_df[new_attrs]
#     print(temp_df.columns)
    
    temp_df = temp_df.rename(columns = {'rolAvg_network_polarization': 'network_polarization', 'rolAvg_net_polarization2': 'network_polarization2', 'rolAvg_network_homophily':'network_homophily', 
                   'rolAvg_network_homophily2':'network_homophily2','rolAvg_network_polarity': 'network_polarity', 'rolAvg_user_satisfaction': 'user_satisfaction'})
    
    return temp_df

attr = ['network_polarization', 'net_polarization2', 'network_homophily', 'network_homophily2', 'network_polarity', 'user_satisfaction']
```

```python
rollAvg_df_se_none = get_rolling_averaged_df(df_avg_se_none, attr, 100)
rollAvg_df_se_high = get_rolling_averaged_df(df_avg_se_high, attr, 100)
rollAvg_df_se_med = get_rolling_averaged_df(df_avg_se_med, attr, 100)
rollAvg_df_se_low = get_rolling_averaged_df(df_avg_se_low, attr, 100)

rollAvg_df_se_none.to_csv('results/final_results/SE_avg/res_rolavg_none.csv')
rollAvg_df_se_high.to_csv('results/final_results/SE_avg/res_rolavg_high.csv')
rollAvg_df_se_med.to_csv('results/final_results/SE_avg/res_rolavg_med.csv')
rollAvg_df_se_low.to_csv('results/final_results/SE_avg/res_rolavg_low.csv')
```

```python
def get_sim_run_sum(df, sim_set):
    
    col = ['unique_authors_count'] + list(df.columns[4:8])
    
    temp_df = pd.DataFrame([[len(set(df.author_id)), df['num_of_agents_received'].sum(), df['num_of_agents_not_received'].sum(), df['num_of_spreader_agents'].sum(), df['num_of_disinterested_agents'].sum()]], 
                            columns = col)
    
    total_agents = temp_df['num_of_agents_received'] + temp_df['num_of_agents_not_received']
    temp_df['unique_authors_count'] = (temp_df['unique_authors_count'] * 100)/4039
    temp_df['num_of_agents_received'] = (temp_df['num_of_agents_received'] * 100)/total_agents
    temp_df['num_of_agents_not_received'] = (temp_df['num_of_agents_not_received'] * 100)/total_agents
    temp_df['num_of_spreader_agents'] = (temp_df['num_of_spreader_agents'] * 100)/total_agents
    temp_df['num_of_disinterested_agents'] = (temp_df['num_of_disinterested_agents'] * 100)/total_agents
    temp_df['sim_set'] = sim_set
#     temp_df = temp_df.drop(['num_of_agents_not_received'], axis = 1)
    
    #print(temp_df)
    
    return temp_df
```

```python
sum_df = pd.DataFrame()

sum_df = pd.concat((sum_df, get_sim_run_sum(df_avg_tol_low, 'TOL-low')), axis = 0) 
sum_df = pd.concat((sum_df, get_sim_run_sum(df_avg_tol_med, 'TOL-med')), axis = 0) 
sum_df = pd.concat((sum_df, get_sim_run_sum(df_avg_tol_high, 'TOL-high')), axis = 0) 
sum_df = pd.concat((sum_df, get_sim_run_sum(df_avg_se_none, 'SE-none')), axis = 0) 
sum_df = pd.concat((sum_df, get_sim_run_sum(df_avg_se_low, 'SE-low')), axis = 0) 
sum_df = pd.concat((sum_df, get_sim_run_sum(df_avg_se_med, 'SE-med')), axis = 0) 
sum_df = pd.concat((sum_df, get_sim_run_sum(df_avg_se_high, 'SE-high')), axis = 0)
sum_df = sum_df.round(2)
```

```python
def get_user_stats(df, sim_setup):
    
    num_agents = 4039
    
    neutral_sat_user_count, sat_user_count, unsat_user_count = df[df['user_satisfaction'] == 0].shape[0], df[df['user_satisfaction'] > 0].shape[0], df[df['user_satisfaction'] < 0].shape[0]
    low_active, med_active, high_active = df[df['user_activity'] <= 0.75].shape[0], df[(df['user_activity'] > 0.75) & (df['user_activity'] < 0.90)].shape[0], df[df['user_activity'] >= 0.90].shape[0]
    neutral_users, med_pol_users, highly_pol_users = df[df['pol_inclination'] == 0].shape[0], df[(df['pol_inclination'] > -0.1) & (df['pol_inclination'] < 0.1)].shape[0], df[(df['pol_inclination'] <= -0.1) | (df['pol_inclination'] >= 0.1)].shape[0]
    
    cols = ['nsat_users', 'sat_users', 'unsat_users', 'low_activity', 'med_activity', 'high_activity', 'npolarized', 'mpolarized', 'hpolarized']
    temp_df = pd.DataFrame([[neutral_sat_user_count, sat_user_count, unsat_user_count, low_active, med_active, high_active, neutral_users, med_pol_users, highly_pol_users]], columns = cols)
    temp_df = (temp_df * 100)/num_agents
    temp_df['sim_setup'] = sim_setup
    
    return temp_df
```

```python
#temp_df = pd.concat((temp_df, temp_df1_i), axis = 0)
temp_df = pd.DataFrame()

temp_df =  pd.concat((temp_df, get_user_stats(initial_data, 'Initial')), axis = 0)
temp_df = pd.concat((temp_df, get_user_stats(df_avg_tol_low_fd, 'TOL-low')), axis = 0)
temp_df = pd.concat((temp_df, get_user_stats(df_avg_tol_med_fd, 'TOL-med')), axis = 0)
temp_df = pd.concat((temp_df, get_user_stats(df_avg_tol_high_fd, 'TOL-high')), axis = 0)

temp_df = pd.concat((temp_df, get_user_stats(df_avg_se_none_fd, 'SE-none')), axis = 0)
temp_df = pd.concat((temp_df, get_user_stats(df_avg_se_low_fd, 'SE-low')), axis = 0)
temp_df = pd.concat((temp_df, get_user_stats(df_avg_se_med_fd, 'SE-med')), axis = 0)
temp_df = pd.concat((temp_df, get_user_stats(df_avg_se_high_fd, 'SE-high')), axis = 0)
```
